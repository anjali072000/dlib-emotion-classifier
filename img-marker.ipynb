```python
import cv2
import dlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

```python
img = cv2.imread('test/t1.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
rects = detector(gray, 1)

for rect in rects:
    shape = predictor(gray, rect)

    shape_np = np.zeros((68, 2), dtype="int")

    for i in range(0, 68):
        shape_np[i] = (shape.part(i).x, shape.part(i).y)
    shape = shape_np
    
    for i, (x, y) in enumerate(shape):
        cv2.circle(img, (x, y), 2, (0, 255, 0), -1)
```

```python
plt.imshow(img)
```

```python
rects = detector(gray, 1)

for rect in rects:
    shape = predictor(gray, rect)

    shape_np = np.zeros((68, 2), dtype="int")

    for i in range(0, 68):
        shape_np[i] = (shape.part(i).x, shape.part(i).y)
    shape = shape_np
    
    for i, (x1, y1) in enumerate(shape[:9]):
        for j, (x2, y2) in enumerate(shape[9:17]):
            cv2.line(img, (x1,y1), (x2,y2), (0, 255, 0), 1)

    for i, (x1, y1) in enumerate(shape[0:17]):
        cv2.line(img, (x1,y1), (shape[i+17]), (0, 255, 0), 1)
        cv2.line(img, (shape[i+17]), (shape[17-i]), (0, 255, 0), 1)
```
